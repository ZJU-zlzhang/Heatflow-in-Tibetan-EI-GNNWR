```python
def get_nn_dis(sample_coords, ref_coords_x, ref_coords_y):
    dis_list = []
    for i in range(sample_coords.shape[0]):
        sample_coords_x = sample_coords[i][0]
        sample_coords_y = sample_coords[i][1]
        dis = np.sqrt(np.square(ref_coords_x - sample_coords_x) + np.square(ref_coords_y - sample_coords_y))
        dis_list.append(dis)
    return np.array(dis_list)
```

```python
import tensorflow as tf

feature_num = 12

# common part ----- dis
def dis_in_gnnwr(input_dis):
    dis = input_dis

    # dis = tf.keras.layers.Dense(1000, activation='relu', kernel_initializer=tf.initializers.HeNormal(),
    #                             kernel_regularizer=tf.keras.regularizers.l2(0.01))(dis)
    # dis = tf.keras.layers.Dense(512, activation='relu', kernel_initializer=tf.initializers.HeNormal(),
    #                             kernel_regularizer=tf.keras.regularizers.l2(0.01))(dis)
    dis = tf.keras.layers.Dense(256, activation='relu', kernel_initializer=tf.initializers.HeNormal(),
                                kernel_regularizer=tf.keras.regularizers.l2(0.01))(dis)
    dis = tf.keras.layers.Dense(64, activation='relu', kernel_initializer=tf.initializers.HeNormal(),
                                kernel_regularizer=tf.keras.regularizers.l2(0.01))(dis)

    dis = tf.keras.layers.Dense(feature_num)(dis)
    return dis


def gnnwr(beta_ols):
    input_dis = tf.keras.layers.Input(shape=dis_train.shape[1:])
    inputs_x = tf.keras.layers.Input(shape=x_train.shape[1:])

    dis = dis_in_gnnwr(input_dis)

    weights = []
    for i in range(feature_num):
        weights.append(dis[:, i])

    y_pred = weights[0] * beta_ols[0][0]
    for i in range(1, feature_num):
        y_pred += weights[i] * beta_ols[i][0] * inputs_x[:, i-1]

    # y_bin = tf.keras.layers.Lambda(lambda z: tf.nn.sigmoid(z))(y_pred)
    # y_bin = tf.keras.layers.Lambda(lambda z: my_round(z))(y_bin)

    return tf.keras.Model(inputs=(input_dis, inputs_x), outputs=y_pred)
```

```python
import os
import datetime

TIMESTAMP = datetime.datetime.now().strftime("%Y_%m_%d_%H_%M_%S").replace("'", "")

nn_logdir = 'F:/0实验室工作相关/0paper/heatflow/exp/gnnwrzzl/heatflow'
if not os.path.exists(nn_logdir):
    os.mkdir(nn_logdir)
nn_output_model_path = os.path.join(nn_logdir, "gnnwr_model_{}.h5".format(TIMESTAMP))
nn_optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)


def nn_scheduler(epoch, lr):
    if epoch < 60:
        return 0.001
    else:
        return lr * tf.math.exp(-0.02)

nn_callbacks = [
    tf.keras.callbacks.TensorBoard(nn_logdir),
    tf.keras.callbacks.ModelCheckpoint(nn_output_model_path, save_best_only=True),
    tf.keras.callbacks.LearningRateScheduler(nn_scheduler),
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=1e-4, patience=40, restore_best_weights=True)
]
```

```python
from sklearn.model_selection import KFold
from pysal.spreg import OLS
from sklearn import metrics
import os

NN_BATCH_SIZE = 16
NN_BUFFER_SIZE = 100
TRAIN_SIZE = 0.9
N_SPLIT = int(1/(1 - TRAIN_SIZE))
# N_SPLIT = 10
RANDOM_STATE = 8

for kfold, (train, valid) in enumerate(KFold(n_splits=N_SPLIT, random_state=RANDOM_STATE, shuffle=True).split(x_train_and_valid, y_train_and_valid)):
    print(f'Cross Validation {kfold} start...')
    # clear the session
    tf.keras.backend.clear_session()

    # prepare data
    dis_train = get_nn_dis(data_coords_train_and_valid[train], data_coords_x, data_coords_y)
    dis_valid = get_nn_dis(data_coords_train_and_valid[valid], data_coords_x, data_coords_y)
    x_train = x_train_and_valid[train]
    x_valid = x_train_and_valid[valid]
    y_train = y_train_and_valid[train]
    y_valid = y_train_and_valid[valid]
    nn_train_input = tf.data.Dataset.from_tensor_slices((dis_train, x_train))
    nn_train_output = tf.data.Dataset.from_tensor_slices(y_train)
    nn_train = tf.data.Dataset.zip((nn_train_input, nn_train_output))
    nn_train = nn_train.shuffle(NN_BUFFER_SIZE).batch(NN_BATCH_SIZE)

    nn_valid_input = tf.data.Dataset.from_tensor_slices((dis_valid, x_valid))
    nn_valid_output = tf.data.Dataset.from_tensor_slices(y_valid)
    nn_valid = tf.data.Dataset.zip((nn_valid_input, nn_valid_output))
    nn_valid = nn_valid.batch(NN_BATCH_SIZE)

    # construct model
    ols = OLS(y_train_and_valid[train], x_train_and_valid[train], name_y = 'HF', name_x = ['MeanCurv', 'LAB','Moho','Topo','Tectonics','Ridge','Transform','Trench','YoungRift','Volcanos','Sus','Bz5'])
    beta_ols = ols.betas
    print(f'beta_0 in ols is {beta_ols[0][0]}')

    # calling the model and compile it
    gnnwr_model = gnnwr(beta_ols)
    gnnwr_model.compile(optimizer=nn_optimizer, loss='mean_squared_error')

    # run the model
    nn_history = gnnwr_model.fit(nn_train, validation_data=nn_valid, epochs=3000)

    # predict test
    # list[metric]
    nn_y_test = gnnwr_model.predict((dis_test, x_test))
    # print(f'y test sample {kfold} R2:', metrics.r2_score(y_test, nn_y_test))
    # print(f'y test sample {kfold} RMSE', np.sqrt(metrics.mean_squared_error(y_test, nn_y_test)))
    # print('')

    with open(f'F:/0实验室工作相关/0paper/heatflow/exp/gnnwrzzl/heatflow_nn_callbacks/score_result{kfold}.txt','a') as file:
        print(f'y test sample {kfold} R2:', metrics.r2_score(y_test, nn_y_test), file=file)
        print(f'y test sample {kfold} RMSE', np.sqrt(metrics.mean_squared_error(y_test, nn_y_test)), file=file)

    pd.DataFrame(beta_ols).to_csv(f'F:/0实验室工作相关/0paper/heatflow/exp/gnnwrzzl/heatflow_nn_callbacks/beta_ols_cv_lr0.001_test1{kfold}.csv', index=False)
    gnnwr_model.save_weights(f'F:/0实验室工作相关/0paper/heatflow/exp/GTCNNWR/heatflow_cv10_test1_wg_{kfold}.h5')
    # gnnwr_model.save(f'F:/0实验室工作相关/0paper/heatflow/exp/GTCNNWR/heatflow_cv10_{kfold}.h5')
    print(f'Cross Validation {kfold} finished.')
```
